* Importing all examples
* For each example, fitting `Model` and `ModelMeans` (with default objective function `MLW`) and inspecting results

```python
from semopy import Model, ModelMeans
from semopy.examples import *
```

```python
models_info = (univariate_regression,
               univariate_regression_many,
               multivariate_regression,
               holzinger39,
               political_democracy,
               example_article,
               # example_rf
               )
```

```python
for info in models_info:
    data, desc = info.get_data(), info.get_model()
    print(f"{'-'*20}\n{desc}\n")
    for model_class in (Model, ModelMeans):
        m = model_class(desc)
        m.fit(data)  # obj: MLW
        print(m.inspect())
```

```python
art_data, art_desc = example_article.get_data(), example_article.get_model()
print(art_desc)
```

```python
art_model = Model(art_desc)
art_model.fit(art_data)
print(art_model.inspect())
```

```python
from semopy import semplot
plot = semplot(art_model, "art_model.pdf")
```
